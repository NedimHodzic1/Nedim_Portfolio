```python
import numpy as np
from numpy.random import seed
from numpy.random import normal
import matplotlib.pyplot as plt
import statistics as stats
from scipy.stats import sem
from scipy.stats import iqr

M1 = np.array([[1, 3, 5], [4, 12, 8]])
M2 = np.array([[1, 5, 9], [3, 4, 15]])

#Section 1 Part 1
M1+M2

#Section 1 Part 2
M1-M2

#Section 1 Part 3A
np.matmul(M1, M2.T)

#Section 1 Part 3B
np.matmul(M2, M1.T)

#Section 1 Part 3C
# They are not the same (values 90 & 136 swap places)

#Section 1 Part 4A
V1 = [[2, 6, 7]]
M3 = np.append(M1, V1, axis=0)

#Section 1 Part 4B - this is not invertable
#np.linalg.inv(M3)

#Section 1 Part 4C - this does not have a determinant
#np.linalg.det(M3)

#Section 1 Part 4D
V2 = [[2, 0, 7]]
M4 = np.append(M1, V2, axis=0)

#Section 1 Part 4E
values, vectors = np.linalg.eig(M4)
f'Eigenvalues of M4:\n{values}\n'
f'Eigenvectors of M4:\n{(vectors)}'

#Section 2: Probability

#Section 2 Part 1A:
np.random.seed(155)
x1 = np.random.normal(loc=15, scale=2, size=40)
print(x1)

#Section 2 Part 1B:
np.random.seed(155)
x2 = np.random.normal(loc=17, scale=1.5, size=40)
print(x2)

#Section 2 Part 1C:
plt.hist(x1, bins=30, alpha=0.5, color='red')
plt.hist(x2, bins=30, alpha=0.5, color='blue')
plt.title("Histogram: x1 and x2")
plt.show()

#Section 2 Part 2: Plot the distribution
x3 = x2 - x1
print(x3)
plt.hist(x3, bins=30, alpha=0.5, color='red')
plt.title("Histogram: x3")
plt.show()

#Section 2 Part 2: Histogram of the differences
plt.boxplot(x3, showmeans=True, vert=False)
plt.title("x3 Distributions")
plt.show()

#Section 3: Statistics

#Section 3 Part 1
x1mean = np.mean(x1)
x2mean = np.mean(x2)
x3mean = np.mean(x3)
x1median= np.median(x1)
x2median= np.median(x2)
x3median= np.median(x3)
x1mode= stats.mode(x1)
x2mode= stats.mode(x2)
x3mode= stats.mode(x3)

print('x1mean = ', x1mean)
print('x2mean = ', x2mean)
print('x3mean = ', x3mean)
print('x1median = ', x1median)
print('x2median = ', x2median)
print('x3median = ', x3median)
print('x1mode = ', x1mode)
print('x2mode = ', x2mode)
print('x3mode = ', x3mode)

#Section 3 Part 1A: The mode is not what you would expect due to m3's mode being the first value in its dataset since all values are unqiue

#Section 3 Part 1b: The new rounded mode is 1.688
rounded_x3mean= np.round(x3mean, decimals=3)
rounded_x3median= np.round(x3median, decimals=3)
rounded_x3mode= np.round(x3mode, decimals=3)

print('rounded_x3mean = ', rounded_x3mean)
print('rounded_x3median = ', rounded_x3median)
print('rounded_x3mode = ', rounded_x3mode)

#Section 3 Part 2
x1_variance= np.var(x1)
x2_variance= np.var(x2)
x3_variance= np.var(x3)
#need ddof=1 here because it returns sample standard deviation, np.std defaults to population standard deviation
x1_sample_standard_deviation = np.std(x1, ddof =1)
x2_sample_standard_deviation = np.std(x2, ddof =1)
x3_sample_standard_deviation = np.std(x3, ddof =1)
#The larger the standard error of the mean, the more spread out values are around the mean in a dataset.
x1_standard_error = sem(x1)
x2_standard_error = sem(x2)
x3_standard_error = sem(x3)

print('x1_variance = ', x1_variance)
print('x2_variance = ', x2_variance)
print('x3_variance = ', x3_variance)
print('x1_sample_standard_deviation = ', x1_sample_standard_deviation)
print('x2_sample_standard_deviation = ', x2_sample_standard_deviation)
print('x3_sample_standard_deviation = ', x3_sample_standard_deviation)
print('x1_standard_error = ', x1_standard_error)
print('x2_standard_error = ', x2_standard_error)
print('x3_standard_error = ', x3_standard_error)

#Section 3 Part 3A
#ptp finds the range of values (max-min) along an axis
#the name of the function comes from the acronym for (peak to peak)
x1_range = np.ptp(x1)
x2_range = np.ptp(x2)

print('x1_range = ', x1_range)
print('x2_range = ', x2_range)

#Section 3 Part 3B
#numpy.quantile - : Compute the qth quantile of the given data (array elements) along the specified axis.
#I'm not sure what quantile we have to compute this for x1 and x2 for, did it for 75,50,25

Q1_75th_quantile = np.quantile(x1, .75)
Q2_75th_quantile = np.quantile(x2, .75)
Q1_50th_quantile = np.quantile(x1, .50)
Q2_50th_quantile = np.quantile(x2, .50)
Q1_25th_quantile = np.quantile(x1, .25)
Q2_25th_quantile = np.quantile(x2, .25)

print('Q1_75th_quantile = ', Q1_75th_quantile)
print('Q2_75th_quantile = ', Q2_75th_quantile)
print('Q1_50th_quantile = ', Q1_50th_quantile)
print('Q2_50th_quantile = ', Q2_50th_quantile)
print('Q1_25th_quantile = ', Q1_25th_quantile)
print('Q2_25th_quantile = ', Q2_25th_quantile)

#Section 3 Part 3C
#The interquartile range (IQR) is the difference between the 75th and 25th percentile of the data.
IQR_x1 = iqr(x1)
IQR_x2 = iqr(x2)

print('IQR_x1 = ', IQR_x1)
print('IQR_x2 = ', IQR_x2)
```
