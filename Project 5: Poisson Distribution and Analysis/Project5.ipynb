```python
#Nedim Hodzic
#Exercise 03 : T54-574

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
import statistics
from scipy.stats import sem
import psycopg2
from scipy.stats.stats import pearsonr
import scipy.stats as stats

#Section 1: Sample Size 

#Section 1 Part 1A - x1
#5 observations from a random normal distribution with a mean of 15 and standard deviation of 2
#Use seed (205) to make sure results are the same
np.random.seed(205)
#loc = mean, scale = standard deviation, size = observations
x1 = np.random.normal(loc=15, scale=2, size=5)
print('x1 = ', x1)

#Section 1 Part 1B - x2
#25 observations from a random normal distribution with a mean of 15 and standard deviation of 2
np.random.seed(205)
x2 = np.random.normal(loc=15, scale=2, size=25)
print('x2 = ', x2)

#Section 1 Part 1C - x3
#125 observations from a random normal distribution with a mean of 15 and standard deviation of 2
np.random.seed(205)
x3 = np.random.normal(loc=15, scale=2, size=125)
print('x3 = ', x3)

#Section 1 Part 1D - x4
#625 observations from a random normal distribution with a mean of 15 and standard deviation of 2
np.random.seed(205)
x4 = np.random.normal(loc=15, scale=2, size=625)
print('x4 = ', x4)

#Section 1 Part 2
#Use subplots to display the histograms of all four sets of numbers. 
sns.set
fig, axs = plt.subplots(2, 2, figsize=(7, 7))
sns.histplot(data=x1, kde=True, color='olive', label="x1", ax=axs[0, 0])
sns.histplot(data=x2, kde=True, color="skyblue", label="x2", ax=axs[0, 1])
sns.histplot(data=x3, kde=True, color="red", label="x3",ax=axs[1, 0])
sns.histplot(data=x4, kde=True, color="blue", label="x4", ax=axs[1, 1])
plt.show()

#Section 1 Part 3
#Compute the sample means, standard deviations, and standard errors of x1, x2, x3, & x4
x1mean = np.mean(x1)
x2mean = np.mean(x2)
x3mean = np.mean(x3)
x4mean = np.mean(x4)
#Need ddof=1 here because it returns sample standard deviation, np.std defaults to population standard deviation
x1_sample_standard_deviation = np.std(x1, ddof =1)
x2_sample_standard_deviation = np.std(x2, ddof =1)
x3_sample_standard_deviation = np.std(x3, ddof =1)
x4_sample_standard_deviation = np.std(x4, ddof =1)
#The larger the standard error of the mean, the more spread out values are around the mean in a dataset.
x1_standard_error = sem(x1)
x2_standard_error = sem(x2)
x3_standard_error = sem(x3)
x4_standard_error = sem(x4)
print('x1mean = ', x1mean)
print('x2mean = ', x2mean)
print('x3mean = ', x3mean)
print('x4mean = ', x4mean)
print('x1_sample_standard_deviation = ', x1_sample_standard_deviation)
print('x2_sample_standard_deviation = ', x2_sample_standard_deviation)
print('x3_sample_standard_deviation = ', x3_sample_standard_deviation)
print('x4_sample_standard_deviation = ', x4_sample_standard_deviation)
print('x1_standard_error = ', x1_standard_error)
print('x2_standard_error = ', x2_standard_error)
print('x3_standard_error = ', x3_standard_error)
print('x4_standard_error = ', x4_standard_error)

#Section 1 Part 4
#Compare these to each other and the parameters of the distribution they come from. How do they differ? 
#The mean differs in that x3 has the smallest mean at 14.89, while x1 has the largest mean at 15.8. 
#The sample standard deviation differs in that x4's deviation is the highest due to having the greatest amount of data points, while x1 is the smallest.
#The standard error differs in that X4 has the smallest standard error due to having the greatest amount of data points among the data sets.

#Section 2: Poisson Distribution

#Section 2 Part 1
#Using numpy & seed(141), generate a Poisson distribution setting lam=10 and size=1000 
#Note: random.poisson(lam=value, size=value)
np.random.seed(141)
Pdist1 = np.random.poisson(10,1000)

#Section 2 Part 2
#Compute mean and variance of your 1000 random Poisson values 
Pdist_mean1 = np.mean(Pdist1)
Pdist_variance1= np.var(Pdist1)
print('Poisson Distribution 1 Mean = ', Pdist_mean1)
print('Poisson Distribution 1 Variance = ', Pdist_variance1)

#Section 2 Part 3
#Does the lambda = mean = variance for Poisson Distribution 1?  
#Answer: No, the values are relatively similar but are not equal to one another 
#Poisson Distribution 1 Mean =  10.075
#Poisson Distribution 1 Variance =  10.199375

#Section 2 Part 4
np.random.seed(141)
Pdist2 = np.random.poisson(5,1000)
Pdist_mean2 = np.mean(Pdist2)
Pdist_variance2= np.var(Pdist2)
print('Poisson Distribution 2 Mean = ', Pdist_mean2)
print('Poisson Distribution 2 Variance = ', Pdist_variance2)
#Does the lambda = mean = variance for Poisson Distribution 2? 
#Answer: No, the values are relatively similar but are not equal to one another
#Poisson Distribution 2 Mean =  4.998
#Poisson Distribution 2 Variance =  5.033996

#Section 3: Analysis

#Section 3 Part 1: 
#Its late 2005, and your boss at the DVD rental company wants to know how effective his customer 
#promotion program was. He tells you, ‘I want you to give me some descriptive information about how 
#much the customers spent before and after the program started. Were the spending habits similar? Did 
#they differ? Did the program help or make things worse?’
 
#1. What is the outcome? 
# Answer: The outcome is a comprensive report on how customer spending habits reacted to the customer promotion program i.e. customer spend before vs. after program. 
#2. What is the main effect/predictor he wants to understand the impact of? 
# Answer: The main predictor he wants to understand the impact of is the customer promotion program.
#3. What is the hypothesis? 
# Answer: The hypothesis is: "If the customer promotion program is implemented then the spending habits of customers will increase"

#Section 3 Part 2:

#Connect to your local postgres instance from python
conn = psycopg2.connect(database="dvdrental",
                        host="localhost",
                        user="postgres",
                        password = "password")

#Extract desired database
df_dvdrental = pd.io.sql.read_sql("with b4 as (\
    select p.customer_id, sum(p.amount) as Payment_before\
    from rental r\
    left outer join payment p on p.rental_id = r.rental_id\
    where rental_date < cast('2005-07-29' as timestamp) and\
        amount is not null\
    group by p.customer_id),\
    aft as (\
        select p.customer_id, sum(p.amount) as Payment_after\
        from rental r\
        left outer join payment p on p.rental_id = r.rental_id\
        where rental_date >= cast('2005-07-29' as timestamp) and\
            amount is not null\
        group by p.customer_id\
    )\
    select distinct c.customer_id, store_id, first_name, last_name,\
            active, payment_before, payment_after\
    from customer c\
    left outer join b4 r on r.customer_id = c.customer_id\
    left outer join aft a on a.customer_id = c.customer_id\
    where payment_after is not null and payment_before is not null", conn)

print(df_dvdrental)

#Section 3 Part 1 
#Compute summary statistics and create histograms of the payment_before and payment_after variables. (Try using describe()in pandas). 

#Summary statistics
df_dvdrental[['payment_before','payment_after']].describe()

#Plots the distribution of film revenue
df_dvdrental['payment_before'].hist(bins=20, color = 'purple')
plt.title('Payment Before')
plt.show()
df_dvdrental['payment_after'].hist(bins=20, color = 'red')
plt.title('Payment After')
plt.show()

#Section 3 Part 2
#Compute the correlation between these two variables and create a scatterplot 

#Correlation
pearsonr(df_dvdrental['payment_before'], df_dvdrental['payment_after'])

#Plot Scatter Plot
sns.scatterplot(data=df_dvdrental, x="payment_before", y="payment_after")
plt.title('Payment Before & Payement After')
plt.xlabel('Payment Before')
plt.ylabel('Payment After')
plt.show()

#Section 3 Part 3
#Compute a variable which is the difference between the amounts spent before and after the program started:  payment_after – payment_before.  
variable_difference= df_dvdrental['payment_after']- df_dvdrental['payment_before']
print('payment_after - payment_before = ', variable_difference)

#Section 3 Part 4
#Generate a histogram of the difference and conduct a one-sample t-test.
variable_difference.hist(bins=20, color = 'olive')
plt.title('payment_after-payment_before')
plt.show()
#perform one sample t-test
#find the mean. It equals 7.88
print(variable_difference.describe())
#t test, the expected mean for the t test will be 7
stats.ttest_1samp(a=variable_difference, popmean = 0)

#Section 3 Part 5
#Interpret your results
#(statistic=0.9120207558131412, pvalue=0.36212530080551897)
#The pvalue for the DataFrame created for payment_after-payment_before is more than the alpha value of 0.05
#Therefore the deviation from the null hypothesis is not significant
#(The popmean does not significantly deviate from the actual mean of variable_difference which equals 7.88)
```
